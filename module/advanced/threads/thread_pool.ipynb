```python
import time 
import random as rd
import asyncio
from concurrent.futures import ThreadPoolExecutor
```

```python
def run(idx):
    time.sleep(rd.randint(a=1, b=2))
    print(f"This is thread[{idx}]")
    return
```

```python
NUM_THREAD = 10
params = [idx for idx in range(NUM_THREAD*10)]
with ThreadPoolExecutor(max_workers=NUM_THREAD) as executor:
    for result in executor.map(run, params):
        print(f"=====")
    pass
```

```python
async def run_async(idx):
    # time.sleep(rd.randint(a=1, b=2))
    asyncio.sleep(rd.randint(a=1, b=2))
    print(f"This is thread[{idx}]")
    return
```

```python
NUM_THREAD = 10
params = [idx for idx in range(NUM_THREAD*10)]
loop = asyncio.get_running_loop()
with ThreadPoolExecutor(max_workers=NUM_THREAD) as executor:
    for idx in params:
        loop.run_in_executor(executor=executor, func=run_async, args=(idx))
        pass
    pass
```
