```python
import sympy as sp
import sympy.plotting as spplot
import matplotlib.figure as fig
import matplotlib.pyplot as plt
import numpy as np
```

```python
def find_derivate() -> None:
    x = sp.symbols('x')
    # expr = x**2 + x + 1
    expr = 0.15 * x**2 + sp.pi*sp.sin(x)/3
    derivate = sp.Derivative(expr, x).doit(deep=True)
    print(f"{derivate}")
    # sp.init_session()
    sp.plot(derivate)
    return
```

```python
find_derivate()
```
